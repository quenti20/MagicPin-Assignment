```python
def categorize_users(df, high_freq, occasional):
    last_month_dict = {}  # Dictionary to store the last occurrence month of each user

    for index, row in df.iterrows():
        user_id = row['C']
        month = row['month']
        year = row['year']
        
        if user_id not in last_month_dict:
            last_month_dict[user_id] = month
            
        else:
            last_month = last_month_dict[user_id]

            if year == df.loc[index-1, 'year']:
                month_diff = month - last_month
            else:
                month_diff = (month - 1 + (12 - last_month)) % 12
            
            if month_diff > 2:
                occasional.append(user_id)
            else:
                high_freq.append(user_id)

            last_month_dict[user_id] = month
    return list(set(high_freq)),list(set(occasional))
```

```python
high_freq_list = []
occasional_list = []

high_freq_list,occasional_list = categorize_users(df, high_freq_list, occasional_list)

len(high_freq_list)
len(occasional_list)
```

```python
total_users = len(high_freq_list) + len(occasional_list)

# Calculate the percentage of active and inactive users
active_percentage = (len(high_freq_list) / total_users) * 100
inactive_percentage = (len(occasional_list) / total_users) * 100

# Plot the pie chart
labels = ['Active Users', 'Inactive Users']
sizes = [active_percentage, inactive_percentage]
colors = ['#66b3ff', '#99ff99']
explode = (0.1, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Active and Inactive Users')
plt.show()
```

```python
def find_preferred_transaction_type(df, user_list):
    user_preferred_types = {}  # Dictionary to store preferred transaction type for each user

    for user_id in user_list:
        user_transactions = df[df['C'] == user_id]
        transaction_counts = user_transactions['E'].value_counts()
        
        if not transaction_counts.empty:
            max_type = transaction_counts.idxmax()
            user_preferred_types[user_id] = max_type
    
    return user_preferred_types

```

```python
preferred_types_dict = find_preferred_transaction_type(df, high_freq_list)

print("Preferred Transaction Types:")
for user_id, max_type in preferred_types_dict.items():
    print(f"{user_id}: {max_type}")
```

```python
import matplotlib.pyplot as plt
type_counts = pd.Series(preferred_types_dict.values()).value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%.1f%%', colors=plt.cm.Set3.colors)
plt.title('Preferred Transaction Types')
plt.show()
```

```python
#do the same thing with the occasional ones. 
preferred_types_dict_occasional = find_preferred_transaction_type(df, occasional_list)

import matplotlib.pyplot as plt
type_counts = pd.Series(preferred_types_dict_occasional.values()).value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%.1f%%', colors=plt.cm.Set3.colors)
plt.title('Preferred Transaction Types')
plt.show()
```

```python
# Convert 'G' column to float
df['G'] = df['G'].astype(float)

# Calculate 25th and 75th percentiles
percentile_25 = df['G'].quantile(0.25, interpolation='lower')
percentile_75 = df['G'].quantile(0.75, interpolation='lower')

def get_currency_preference(df, percentile_25, percentile_75):
    quarter_dict = {}
    three_quarter_dict = {}
    
    for index, row in df.iterrows():
        user_id = row['C']
        amount = row['G']
        currency = row['F']
        
        if user_id not in quarter_dict and amount <= percentile_25:
            quarter_dict[user_id] = currency
        elif user_id not in three_quarter_dict and amount >= percentile_75:
            three_quarter_dict[user_id] = currency
    
    return quarter_dict, three_quarter_dict
```

```python
quarter_dict, three_quarter_dict = get_currency_preference(df, percentile_25, percentile_75)
```

```python
def find_preferred_currency_type(df, three_quarter_list):
    user_currency_types = {}  # Dictionary to store preferred transaction type for each user

    for user_id in user_list:
        user_transactions = df[df['C'] == user_id]
        currency_counts = user_transactions['F'].value_counts()
        
        if not currency_counts.empty:
            max_type = transaction_counts.idxmax()
            user_currency_types[user_id] = max_type
    
    return user_currency_types
```

```python
import matplotlib.pyplot as plt
quarter_type_counts = pd.Series(quarter_dict.values()).value_counts()
three_quarter_type_counts = pd.Series(three_quarter_dict.values()).value_counts()

quarter_type_counts
```
